```python
import pandas as pd
import numpy as np
import math
import sys
np.set_printoptions(threshold=sys.maxsize, linewidth=1000, precision=3)

years = [2562, 2563, 2564, 2565, 2566]
# Create a new structure for Dividend Yield (%)
dividend_yield_data = {year: {} for year in years}
price_yield_data = {year: {} for year in years}
combined_data = {year: {} for year in years}
num_years = len(years) # 5

for ticker, data in stock_info.items():
    previous_price = None
    for index, year in enumerate(years): 
        price = data["prices"][num_years - 1 - index]  # Access prices in reverse order
        dividend = data["dividends"][num_years - 1 - index]  # Access dividends in reverse order
        dividend_yield = (dividend / price) * 100 if price != 0 else 0
        
        dividend_yield_data[year][ticker] = dividend_yield
        
        # Calculate Price Yield
        if previous_price is not None:
            price_change = ((price - previous_price) / previous_price) * 100
        else:
            price_change = 0  # No change for the first year
        
        price_yield_data[year][ticker] = price_change
        
        # Calculate Combined Data
        combined_data[year][ticker] = dividend_yield + price_change

        # Update previous price for next iteration
        previous_price = price

# Convert the data into a DataFrame for better visualization
dividend_yield_df = pd.DataFrame(dividend_yield_data).T
price_yield_df = pd.DataFrame(price_yield_data).T
combined_df = pd.DataFrame(combined_data).T

# Convert the DataFrame to a NumPy array without the year column
dividend_yield_matrix = dividend_yield_df.to_numpy()
price_yield_matrix = price_yield_df.to_numpy()
combined_matrix = combined_df.to_numpy()

# Print results
print("Dividend Yield DataFrame:")
print(dividend_yield_df.to_string())
print("\nPrice Yield DataFrame:")
print(price_yield_df.to_string())
print("\nCombined DataFrame (Dividend + Price Yield):")
print(combined_df.to_string())
print("\nShapes:")
print("Dividend Yield Matrix Shape:", dividend_yield_matrix.shape)
print("Price Yield Matrix Shape:", price_yield_matrix.shape)
print("Combined Matrix Shape:", combined_matrix.shape)
```

# Covariance matrix & Bordered Hessian matrix

```python
import numpy as np
from joblib import Parallel, delayed
from tqdm import tqdm

def calculate_covariance_matrix(selected_matrix):
    """Calculate the covariance matrix for the selected stocks."""
    mean_values = np.mean(selected_matrix, axis=0)
    centered_data = selected_matrix - mean_values
    covariance_matrix = (centered_data.T @ centered_data) / (selected_matrix.shape[0] - 1)
    return covariance_matrix, mean_values

def calculate_bordered_hessian(covariance_matrix, mean_values):
    """Calculate the bordered Hessian matrix."""
    n = covariance_matrix.shape[0]
    expected_returns = mean_values
    negative_identity = -np.identity(n)
    combined_matrix = np.hstack((
        covariance_matrix, 
        np.where(negative_identity == -0, 0, negative_identity), 
        np.ones((n, 1)), 
        expected_returns.reshape(-1, 1)
    ))

    vertical_one = np.ones((1, n))
    vertical_zero = np.zeros((1, n+2))
    combined_vertical = np.hstack((vertical_one, vertical_zero))
    combined_matrix = np.vstack((combined_matrix, combined_vertical))

    vertical_expected = expected_returns.reshape(1, -1)
    vertical_zero = np.zeros((1, n+2))
    combined_vertical = np.hstack((vertical_expected, vertical_zero))
    combined_matrix = np.vstack((combined_matrix, combined_vertical))

    return combined_matrix

def process_combination(combination, yield_matrix, stocks, verbose=True):
    selected_indices = [stocks.index(stock) for stock in combination]
    selected_matrix = yield_matrix[:, selected_indices]

    # Step 1: Calculate the covariance matrix and mean values
    covariance_matrix, mean_values = calculate_covariance_matrix(selected_matrix)

    # Step 2: Calculate the Bordered Hessian matrix
    bordered_hessian = calculate_bordered_hessian(covariance_matrix, mean_values)
    bordered_hessian = calculate_bordered_hessian(covariance_matrix, mean_values)

    return combination, covariance_matrix, bordered_hessian
```

# Solve Equation

```python
import itertools
import numpy as np
from joblib import Parallel, delayed
from tqdm import tqdm
# Define a function to compute solutions for a single combination and ci_case
def compute_solutions(combination, covariance_matrix, H, ci_case, r):
    n = H.shape[0] - 2  # Number of stocks (n = 5 in this case)
    cols_to_keep = []

    # Determine which rows/columns to keep based on the current ci_case
    for i, ci in enumerate(ci_case):
        if ci == 1:  # c_i > 0, lambda_i = 0
            cols_to_keep.append(i)  # Keep c_i column
        else:  # c_i = 0, lambda_i > 0
            cols_to_keep.append(i + n)  # Keep lambda_i column
    # Add rows/columns for the additional constraints (1 and E parts)
    cols_to_keep += [H.shape[1]-2, H.shape[1] -1]  # Keep columns for 1 and expected returns (E)
    cols_to_keep.sort() #!! 2 days to solve this bug (forgot to sort) T^T
    
    # Create the reduced matrix H_reduced
    H_reduced = H[:, cols_to_keep]

    # Create the b vector
    b = np.zeros((H.shape[0], 1))
    b[-2] = 1  # Constraint on weights sum
    b[-1] = r  # Required return 'r'
    
    # Try to solve the reduced system
    try:
        x_reduced = np.linalg.solve(H_reduced, b)
        return (combination,covariance_matrix, ci_case, x_reduced, H, H_reduced, b)
    except np.linalg.LinAlgError:
        return None  # Return None if there is an error
```

```python
import pickle
import os
def save_results(filename, data):
    # Create directory if it doesn't exist
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    
    with open(filename, 'wb') as f:
        pickle.dump(data, f)

def load_results(filename):
    with open(filename, 'rb') as f:
        return pickle.load(f)
    
solutions_file = '../out/raw_data/valid_solutions10Percentv1.pkl'
r = 10  # Required return of 10%
valid_solutions = {}
# Check if results file exists
if os.path.exists(solutions_file):
    print(f"Loading results from {solutions_file}...")
    valid_solutions = load_results(solutions_file)
else:
    print("Running the computation...")
    combinations_ci = list(itertools.product([0, 1], repeat=5))
    tasks = list(itertools.product(combinations_ci, H_dividend_yield__matrices.items()))

    # Parallel computation of results
    results = Parallel(n_jobs=-1)(
        delayed(compute_solutions)(
            combination,
            covariance_matrix,
            H,
            ci_case,
            r
        )
        for ci_case, (combination, (H, covariance_matrix)) in tqdm(tasks, desc="Processing tasks", total=len(tasks))
    )
    # Filter out None results and organize them into the solutions dictionary
    print("Filtering out None results...")
    solutions = {}
    for result in results:
        if result:
            combination, covariance_matrix, ci_case, x_reduced, H, H_reduced, b = result
            if combination not in solutions:
                solutions[combination] = []  # Initialize an empty list for this combination
            solutions[combination].append((covariance_matrix, ci_case, x_reduced, H, H_reduced, b))
    print("Successfully filtered out None results!")
    
    # Filter to get only valid solutions (lampda>=0, c>=0)
    print("Filtering valid solutions... (lampda>=0, c>=0)")
    for combination, solution_list in solutions.items():
        for covariance_matrix, ci_case, x_reduced, H, H_reduced, b in solution_list:
            checker = {}
            x_values = x_reduced.flatten()  # Convert the array to a 1D array
            mu_values = {}
            mu_values['mu1'] = x_values[-1]
            mu_values['mu2'] = x_values[-2]
            increment = 0
            for i,case in enumerate(ci_case):
                if case == 1:
                    checker[f'c{i+1}'] = x_values[increment]
                    increment += 1
            for i,case in enumerate(ci_case):
                if case == 0:
                    checker[f'lambda{i+1}'] = x_values[increment]
                    increment += 1

            if all(value>=0 for value in checker.values()):
                if combination not in valid_solutions:
                    valid_solutions[combination] = []
                valid_solutions[combination].append((covariance_matrix, ci_case, x_reduced, checker, mu_values))
    print(f"Total number of valid solutions: {len(valid_solutions)}")
    print(f"Saving valid solutions to {solutions_file}...")
    save_results(solutions_file, valid_solutions)
    print(f"Successfully saved valid solutions!")
```

```python
import pickle
import os
def save_results(filename, data):
    # Create directory if it doesn't exist
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    
    with open(filename, 'wb') as f:
        pickle.dump(data, f)

def load_results(filename):
    with open(filename, 'rb') as f:
        return pickle.load(f)
    
solutions_file = '../out/raw_data/valid_solutions10Percentv2.pkl'
r = 10  # Required return of 10%
valid_solutions = {}
# Check if results file exists
if os.path.exists(solutions_file):
    print(f"Loading results from {solutions_file}...")
    valid_solutions = load_results(solutions_file)
else:
    print("Running the computation...")
    combinations_ci = list(itertools.product([0, 1], repeat=5))
    tasks = list(itertools.product(combinations_ci, price_yield_H_matrices.items()))

    # Parallel computation of results
    results = Parallel(n_jobs=-1)(
        delayed(compute_solutions)(
            combination,
            covariance_matrix,
            H,
            ci_case,
            r
        )
        for ci_case, (combination, (H, covariance_matrix)) in tqdm(tasks, desc="Processing tasks", total=len(tasks))
    )
    # Filter out None results and organize them into the solutions dictionary
    print("Filtering out None results...")
    solutions = {}
    for result in results:
        if result:
            combination, covariance_matrix, ci_case, x_reduced, H, H_reduced, b = result
            if combination not in solutions:
                solutions[combination] = []  # Initialize an empty list for this combination
            solutions[combination].append((covariance_matrix, ci_case, x_reduced, H, H_reduced, b))
    print("Successfully filtered out None results!")
    
    # Filter to get only valid solutions (lampda>=0, c>=0)
    print("Filtering valid solutions... (lampda>=0, c>=0)")
    for combination, solution_list in solutions.items():
        for covariance_matrix, ci_case, x_reduced, H, H_reduced, b in solution_list:
            checker = {}
            x_values = x_reduced.flatten()  # Convert the array to a 1D array
            mu_values = {}
            mu_values['mu1'] = x_values[-1]
            mu_values['mu2'] = x_values[-2]
            increment = 0
            for i,case in enumerate(ci_case):
                if case == 1:
                    
                    checker[f'c{i+1}'] = x_values[increment]
                    increment += 1
            for i,case in enumerate(ci_case):
                if case == 0:
                    checker[f'lambda{i+1}'] = x_values[increment]
                    increment += 1

            if all(value>=0 for value in checker.values()):
                if combination not in valid_solutions:
                    valid_solutions[combination] = []
                valid_solutions[combination].append((covariance_matrix, ci_case, x_reduced, checker, mu_values))
    print(f"Total number of valid solutions: {len(valid_solutions)}")
    print(f"Saving valid solutions to {solutions_file}...")
    save_results(solutions_file, valid_solutions)
    print(f"Successfully saved valid solutions!")
```
